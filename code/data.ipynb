```python
import csv
import pandas as pd
import os

df_price = pd.DataFrame() 

for ix, row in sp500.iterrows():
  try:
    sym, name = row['Symbol'], row['Name']
    csv_fn = os.path.join(folder, f'{sym}.csv')
    print(csv_fn)
    df = pd.read_csv(csv_fn, parse_dates=True, index_col='Date') # 인덱스를 자동으로 datetime으로 설정
    df_price[sym] = df['Close']
  except FileNotFoundError as e:
    print(e)
  except ValueError as e:
    print(e)    
```

```python
df = df_price.iloc[:,:10] # 컬럼이 너무 많아서 10개만 함
```

```python
import numpy as np
```

```python
rets = np.log(df / df.shift(1)) # 수익률 시계열 가장 간단하게 만드는 방법
rets
```

```python
ret = rets.dropna()
```

```python
# 시계열의 Sampling
ret.resample('1w', label='right').last().head()  # 1 week 마지막 날
```

```python
# 시계열의 Sampling
ret.resample('1m', label='right').last().head()  # 1 month 마지막 날
```

```python
import numpy as np
ret.cumsum().apply(np.exp).resample('1m', label='right').last().plot(figsize=(12, 6)) # 월간 수익률
```
