# KOSPI

```python
# KOSPI 종목 전체
df_ks = fdr.StockListing('KOSPI')
df_ks.head()

# S&P 500 symbol list
# sp500 = fdr.StockListing('S&P500')

# stocks = fdr.StockListing('NYSE')   # 뉴욕거래소
# stocks = fdr.StockListing('NASDAQ') # 나스닥
```

```python
len(df_ks) # 950개의 종목
```

```python
df_ks.columns
```

```python
ticker = df_ks[df_ks['Name']=='삼성전자']['Code']
ticker
```

## 2003.06.01~ 2023-06-01

```python
ss = fdr.DataReader(''.join(ticker.values), '2003-06-01', '2023-06-01') #fdr.DataReader('005930', '2003-06-01')
ss.head()
```

```python
ss.tail()
```

```python
ss['Close'].plot()
```

## 2002년~ 2022년

```python
ss = fdr.DataReader(''.join(ticker.values), '2002', '2023') #fdr.DataReader('005930', '2003')
# 2002년~ 2022년
ss.head()
```

```python
ss.tail()
```
