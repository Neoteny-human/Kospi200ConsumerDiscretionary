```python
def multivariate_data(dataset, target, start_index, end_index, history_size, target_size, single_step=False):
    data = []
    labels = []

    start_index = start_index + history_size
    if end_index is None:
        end_index = len(dataset) - target_size

    for i in range(start_index, end_index):
        indices = range(i - history_size, i)
        data.append(dataset[indices])

        if single_step:
            labels.append(target[i + target_size])
        else:
            labels.append(target[i:i + target_size])
    return np.array(data), np.array(labels)
```

```python
BATCH_SIZE = 30 # 미니 배치
# BUFFER_SIZE = 10000 #완벽한 셔플링을 위해서는 데이터 세트의 전체 크기보다 크거나 같은 버퍼 크기가 필요합니다.
EVALUATION_INTERVAL = 4
EPOCHS = 100
```

```python
x_train_single = x_train_single.astype(np.float32)
y_train_single = y_train_single.astype(np.float32)
x_test_single = x_test_single.astype(np.float32)
y_test_single = y_test_single.astype(np.float32)
```

```python
train_data_single = tf.data.Dataset.from_tensor_slices((x_train_single, y_train_single)) # numpy array x(input)를 Dataset으로 변환
train_data_single # <_TensorSliceDataset element_spec=(TensorSpec(shape=(2, 18), dtype=tf.float32, name=None), TensorSpec(shape=(), dtype=tf.float32, name=None))>
train_data_single = train_data_single.cache().batch(BATCH_SIZE).repeat()
# # train_data_single = train_data_single.cache().shuffle(BUFFER_SIZE).batch(BATCH_SIZE).repeat()

test_data_single = tf.data.Dataset.from_tensor_slices((x_test_single, y_test_single))
test_data_single = test_data_single.batch(BATCH_SIZE).repeat()

single_step_model = tf.keras.models.Sequential()
single_step_model.add(tf.keras.layers.LSTM(32, input_shape=x_train_single.shape[-2:]))
single_step_model.add(tf.keras.layers.Dense(1))
single_step_model.compile(optimizer=tf.keras.optimizers.RMSprop(), loss='mean_squared_error')

for x, y in test_data_single.take(1):
    print(single_step_model.predict(x).shape)

single_step_history = single_step_model.fit(train_data_single, epochs=EPOCHS,
                                            steps_per_epoch=EVALUATION_INTERVAL,
                                            validation_data=test_data_single,
                                            validation_steps=50)


def plot_train_history(history, title):
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs = range(len(loss))

    plt.figure()

    plt.plot(epochs, loss, 'b', label='Training loss')
    plt.plot(epochs, val_loss, 'r', label='test loss')
    plt.title(title)
    plt.legend()

    plt.show()

plot_train_history(single_step_history,
                   'Single Step Training and test Loss')
```

```python
# Defind function for Visualization
def create_time_steps(length):
    return list(range(-length, 0))
def show_plot(plot_data, delta, title):
    labels = ['History', 'True Future', 'Model Prediction']
    marker = ['.-', 'rx', 'go']
    time_steps = create_time_steps(plot_data[0].shape[0])
    if delta:
        future = delta
    else:
        future = 0

    plt.title(title)
    for i, x in enumerate(plot_data):
        if i:
            plt.plot(future, plot_data[i], marker[i], markersize=10, label=labels[i])
        else:
            plt.plot(time_steps, plot_data[i].flatten(), marker[i], label=labels[i])
    plt.legend()
    plt.axis('auto')
    plt.xlim([time_steps[0], (future+5)*2])
    plt.xlabel('Time-Step')
    return plt
```

```python
for x, y in test_data_single.take(3):
    plot = show_plot([x[0][:, -1].numpy(), y[0].numpy(),
                      single_step_model.predict(x)[0]], 1,
                     'Single Step Prediction')
    plot.show()
```

# LRP(tensorflow)

```python
import tensorflow as tf

def lrp(model, x, y, epsilon=1e-7):
    # 모델의 입력과 출력을 텐서로 변환
    x_tensor = tf.convert_to_tensor(x, dtype=tf.float32)
    y_tensor = tf.convert_to_tensor(y, dtype=tf.float32)

    with tf.GradientTape() as tape:
        tape.watch(x_tensor)
        predictions = model(x_tensor)  # 모델의 예측 결과 계산
        loss = tf.keras.losses.mean_squared_error(y_tensor, predictions)  # 손실 함수 계산

    # 예측값에 대한 LRP 계산
    gradients = tape.gradient(loss, x_tensor)
    relevance = x_tensor * gradients

    # 정규화
    sum_relevance = tf.reduce_sum(relevance, axis=-1, keepdims=True)
    normalized_relevance = relevance / (sum_relevance + epsilon)

    return normalized_relevance.numpy()


# 사용 예시
# LRP를 적용할 모델과 입력 데이터를 준비한 후 호출합니다.
model = tf.keras.models.load_model('/content/drive/MyDrive/ubion_finance/세진/LSTM_3.h5') # 모델을 새로 불러옴
model.build((None, 2, 18))

# LRP 예시
x_test_batch, y_test_batch = next(iter(test_data_single))

relevance_map = lrp(model, x_test_batch, y_test_batch)
print(relevance_map)
```

위의 결과는 LRP (Layer-wise Relevance Propagation)를 통해 얻은 정규화된 관련성 맵입니다. 각각의 관련성 맵은 입력 데이터의 각 요소와 관련된 중요도를 나타냅니다. 결과는 3차원 배열로 구성되어 있으며, 각 차원은 다음과 같이 해석될 수 있습니다:

- 첫 번째 차원: 입력 시퀀스의 각 타임 스텝에 대한 관련성 맵
- 두 번째 차원: 시퀀스 내 각 특성(feature)에 대한 관련성 맵
- 세 번째 차원: 입력 데이터의 각 피처(feature)에 대한 중요도

각 관련성 맵의 값은 해당 위치의 입력 데이터 요소와 관련된 중요도를 나타냅니다. 값이 양수인 경우 해당 요소가 모델의 출력에 긍정적인 영향을 미친다는 것을 나타내며, 값이 음수인 경우 해당 요소가 모델의 출력에 부정적인 영향을 미친다는 것을 나타냅니다.

따라서, 주어진 결과를 통해 입력 데이터의 각 요소와 피처의 중요도를 확인할 수 있습니다. 예를 들어, 첫 번째 시퀀스의 첫 번째 타임 스텝과 첫 번째 피처에 해당하는 값인 -0.087946은 해당 요소가 모델의 출력에 부정적인 영향을 미친다는 것을 나타냅니다.

결과를 자세히 해석하려면 입력 데이터와 모델의 구조 및 동작에 대한 이해가 필요할 수 있습니다.
