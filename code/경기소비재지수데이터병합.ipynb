```python
import pandas as pd
```

```python
df_1 = pd.read_csv("/content/drive/MyDrive/ubion_finance/Datasets/경기소비재/0809년.csv", encoding='euc-kr')
df_2 = pd.read_csv("/content/drive/MyDrive/ubion_finance/Datasets/경기소비재/1011년.csv", encoding='euc-kr')
df_3 = pd.read_csv("/content/drive/MyDrive/ubion_finance/Datasets/경기소비재/1213년.csv", encoding='euc-kr')
df_4 = pd.read_csv("/content/drive/MyDrive/ubion_finance/Datasets/경기소비재/1415년.csv", encoding='euc-kr')
df_5 = pd.read_csv("/content/drive/MyDrive/ubion_finance/Datasets/경기소비재/1617년.csv", encoding='euc-kr')
df_6 = pd.read_csv("/content/drive/MyDrive/ubion_finance/Datasets/경기소비재/1819년.csv", encoding='euc-kr')
df_7 = pd.read_csv("/content/drive/MyDrive/ubion_finance/Datasets/경기소비재/2021년.csv", encoding='euc-kr')
df_8 = pd.read_csv("/content/drive/MyDrive/ubion_finance/Datasets/경기소비재/2223년.csv", encoding='euc-kr')
```

```python
df = pd.concat([df_1, df_2], axis=0)
df1 = pd.concat([df, df_3], axis=0)
df2 = pd.concat([df1, df_4], axis=0)
df3 = pd.concat([df2, df_5], axis=0)
df4 = pd.concat([df3, df_6], axis=0)
df5 = pd.concat([df4, df_7], axis=0)
df6 = pd.concat([df5, df_8], axis=0)

df6.sort_values(by='일자')
```

```python
df6.to_csv("경기소비재지수_Total.csv", index=False)
```

```python
df = pd.read_csv("/content/drive/MyDrive/ubion_finance/Datasets/경기소비재지수.csv")
```

```python
df = df.rename(columns={'일자': 'Date'})
```

```python
df = df.sort_values(by='Date')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['YM'] = df['Date'].dt.to_period('M')
df = df.groupby('YM').first().reset_index()
df.set_index('Date', inplace=True)
df.drop(columns='YM', inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
# 'Date' 열의 형식을 YYYY-MM으로 변경합니다.
df.index = df.index.to_period('M')
df
```

```python
df = df[['거래량', '종가']]
```

```python
df.to_csv("경기소비재.csv")
```
