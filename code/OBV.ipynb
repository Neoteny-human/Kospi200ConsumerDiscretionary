```python
df = df.set_index('일자')
```

```python
df = df.sort_index()
```

```python
OBV = []
OBV.append(0)
for i in range(1, len(df['종가'])):
  if df['종가'][i] > df['종가'][i-1]:
    OBV.append(OBV[-1] + df['거래량'][i])
  elif df['종가'][i] < df['종가'][i-1]:
    OBV.append(OBV[-1] - df['거래량'][i])
  else:
    OBV.append(OBV[-1])
```

```python
df['OBV'] = OBV
```
