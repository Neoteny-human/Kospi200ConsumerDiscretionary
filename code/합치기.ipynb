```python
import pandas as pd
```

```python
silver = pd.read_csv('vkospi2.csv', encoding='euc_kr')
silver2 = pd.read_csv('vkospi.csv')
```

```python
silver2.columns
```

```python
silver['일자'] = pd.to_datetime(silver['일자'])
silver.set_index('일자', inplace=True)
silver['vkospi'] = silver['종가']
silver = silver[['vkospi']]
silver
```

```python
silver2['Date'] = pd.to_datetime(silver2['Date'], format='%m/%d/%Y')
silver2.set_index('Date', inplace=True)
silver2['vkospi'] = silver2['Price']
silver2 = silver2[['vkospi']]
silver2
```

```python
merged_df = pd.concat([silver2, silver])
merged_df
```

```python
merged_df.to_csv("10년채권.csv")
```
